```python
#How to get the tweets from a given user?
#https://twitter.com/QuotidienSN
import requests
from bs4 import BeautifulSoup
username = "QuotidienSN"
url = "http://www.twitter.com/" + username
print("\n\nDownloading tweets for " + username)
response = None
try:
    response = requests.get(url)
except Exception as e:
    print(repr(e))
    sys.exit(1)
print(response)

soup = BeautifulSoup(response.text, 'html')
tweets = scrapers.get_tweets_data(username, soup)
print(tweets)
```
