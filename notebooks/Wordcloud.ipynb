```python
import numpy as np
import pandas as pd
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import matplotlib.pyplot as plt
```

```python
df= pd.read_csv("../data"+"/Data_cleaned.csv")
```

```python
df.country.unique()
```

```python
df
```

```python
file = open("../data/stopwords_fr.txt", "r") 
french_stopwords=file.read().split("\n")
stop_words=french_stopwords+['article','actucameroun','http','com','ca','faire','pay','egalement','of','the','tout','pays']

```

```python
text = ', '.join(df.text.to_list())
wordcloud = WordCloud(max_font_size=100, max_words=100,
                      background_color="white",collocations=False,
                      stopwords=stop_words).generate(text)
plt.figure(figsize=(20,5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
for country in df.country.unique():
    print(country)
    df_country=df[df["country"]==country]
    text = ', '.join(df_country.text.to_list())
    wordcloud = WordCloud(max_font_size=100, max_words=100,
                          background_color="white",collocations=False,
                          stopwords=stop_words).generate(text)
    plt.figure(figsize=(20,5))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.show()
```
